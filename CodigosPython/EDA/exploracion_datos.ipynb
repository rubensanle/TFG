```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#dataset = pd.DataFrame() #dataset global

inicio = '2001-12-01' #fecha en la que empezamos a descargar datos
fin = '2026-03-01' #fecga en la que dejamos de descargar datos
```

```python
petroleo = yf.download('CL=F', start=inicio, end=fin, interval='1mo')['Close'] #Descargamos los datos de cierre del petroleo
petroleo.info()
petroleo = petroleo.resample("ME").last() #Cogemos el valor de cierre del último día de cada mes, aunque en este caso no es necesario porque el petroleo ya se publica con datos mensuales, pero lo hacemos por consistencia con el resto de variables
petroleo = petroleo.interpolate(method='linear')
petroleo["petroleo_return"] = petroleo.pct_change() #Calculamos el retorno mensual del petroleo y lo guardamos en una nueva columna
#petroleo_retorno = petroleo[["petroleo_return"]]
petroleo = petroleo.reset_index()

#mask = petroleo["petroleo_return"] == 0 #Creamos una mascara para identificar los valores nulos en la columna de retorno del petroleo

#petroleo = petroleo[mask]
petroleo.info()

#Vemos que el return 0 del petroleo, se debe a que los valores son NaN, por lo que vamo a inter`ppolarlos`
```
